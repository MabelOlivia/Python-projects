```python
import pandas as pd
df = pd.read_csv('weather_data.csv', parse_dates =['day'])
df.set_index('day', inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': "no weather record"
})
new_df
```

```python
new_df = df.fillna(method = 'ffill') # method = 'bfill'
new_df
```

```python
new_df = df.fillna(method = 'ffill', limit=1) 
new_df
```

```python
new_df = df.interpolate(method = 'time')
new_df
```

```python
df.head(4).plot.bar(y='temperature', color = 'Blue')
```

```python
new_df.head(4).plot.bar(y='temperature', color = 'Blue')
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how= 'all')
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
